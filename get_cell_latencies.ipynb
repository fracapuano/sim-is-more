```python
import json
with open("searchspaces/nats_arch_index.json", "r") as f:
    network_index = json.load(f)
```

```python
ops = ['skip_connect', 'nor_conv_1x1', 'nor_conv_3x3', 'none', 'avg_pool_3x3']

same_op_nets = [
    "{}~0/{}~0/{}~1/{}~0/{}~1/{}~2".format(op, op, op, op, op, op) for op in ops
]

same_net_index = [
    network_index[s] for s in same_op_nets
]
```

```python
with open("searchspaces/hw_nats_interface_dataset.json", "r") as f: 
    hw_space = json.load(f)
```

```python
dataset, device = "ImageNet16-120", "edgegpu"
[(hw_space[str(i)]["architecture_string"], hw_space[str(i)][dataset][f"{device}_latency"]) for i in same_net_index]
```

```python
import numpy as np
file_name = "nats_cached_metrics.json"
datasets = ["cifar10", "cifar100", "ImageNet16-120"]

output = {
    f"{device}_latency": 
        {
            ops[idx]: np.mean(
                [hw_space[str(i)][dataset][f"{device}_latency"] for dataset in datasets]
            )/len(ops)
                for idx, i in enumerate(same_net_index)
    }
    for device in ["edgegpu", "raspi4", "pixel3", "eyeriss", "fpga"]
}
```

```python
with open(file_name, "w") as f:
    json.dump(output, f, indent=4)
```
